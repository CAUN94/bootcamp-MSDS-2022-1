# Actividad

Crea un ERD para representar la base de datos para una aplicación que rastrea usuarios, libros, libros favoritos de los usuarios y el autor de los libros (asuma que cada libro solo puede tener un autor)

Cada libro debe tener un título y un autor, y cada usuario debe poder guardar una lista de sus libros favoritos. Use draw.io para hacer el modelo de esta BBDD. 

Cuando tenga listo el diagrama debe usar SQL para crear la base de datos, este debe respetar la creación de primary key y foreign key.

Se le pedira que agregue usando sql la siguiente información:
- 5 Usuario
- 5 Libros
- Crear 2 o más registro de usuario y libros favoritos

Se le pedira que realice las siguientes consultas a la base de datos:
- Mostrar a todos los usuarios que tenga al menos un libro favorito
- Mostrar el libro que más se repite en la lista de favoritos
- Modificar el telefono del primer usuario
- Borrar al segundo usuario
