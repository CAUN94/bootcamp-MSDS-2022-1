# Árbol General

Un árbol es una combinación de nodos (también conocidos como vértices) y aristas. Un árbol puede tener cualquier número de nodos y bordes.

Tipos de nodos en un árbol:

Nodo padre: un nodo que tiene uno o más hijos.
Nodo hijo: un nodo que tiene un nodo padre.
Nodo hoja: un nodo que no tiene hijos.

          10
        /    \
       34      89
     /    \ 
    45    50
